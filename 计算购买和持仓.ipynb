```python
import os
import json
import time
import math
import numpy as np

import pandas as pd
from ipysheet import from_dataframe
from ipywidgets import Text, VBox, link
import ipywidgets as widgets
from ipywidgets import Select, IntText, Button, Output, Layout
from ipywidgets import GridspecLayout, AppLayout
from ipysheet import from_dataframe, sheet, cell

from itertools import accumulate
from IPython.display import display, clear_output


# stock = "600273"
# realdata = pd.read_csv('data/' + stock + '.CSV')

# startDate = "2019-11-20"
# endDate = "2019-12-31"

selected = realdata.loc[(realdata['Date'] > startDate) & (realdata['Date'] < endDate)]
    
dates = selected['Date'].values

global inputs
inputs = {}

def caculate_P_high_low(sell_at, buy_at, date):
    global inputs
#     global current_date
#     current_date = date
#     inputs[str(date)] = {}

    inputs[str(date)]['buy_at'] = np.max(buy_at)
    inputs[str(date)]['sell_at'] = np.min(sell_at)
    
#     print(inputs)

# date picker selection callback method
def datepicker_cb(change):
#     print(change)
    global current_date
    current_date = change['new']
    
    global inputs
    inputs[str(change['new'])]={}
    inputs[str(change['new'])]['date'] = change['new']
    inputs[str(change['new'])]['low'] = selected.loc[change['new']==selected['Date']]['Low'].values[0]
    inputs[str(change['new'])]['high'] = selected.loc[change['new']==selected['Date']]['High'].values[0]
    
#     print(inputs)
    for data in day_data:
        if data['date'] == change['new'] :
            caculate_P_high_low(data['high'], data['low'], change['new'])
    
date_picker = Select(options=dates, description='Select Date:', disabled=False, layout=Layout(
    display='flex',
    flex_flow='column',
    border='solid .5px green',
    align_items='stretch',
    height='200px'
))
date_picker.observe(datepicker_cb, names='value')

buy_num = IntText(value=0, description='Buy:', disabled=False)
sell_num = IntText(value=0, description='Sell:', disabled=False)
exe_button = Button(description='Take Order', disabled=False, button_style='', tooltip='Click me')

def on_button_clicked(b):
    global inputs
    global current_date

    if str(current_date) in inputs:
        inputs[str(current_date)]['sell_num'] = sell_num.value
        inputs[str(current_date)]['buy_num'] = buy_num.value
    else:
        inputs[str(current_date)] = {}

    df = pd.DataFrame.from_dict(inputs, orient='index')

    buy_sell_df = check_buy_sell(df)
    remain_df = caculate_remain(buy_sell_df)
    cash_df = cashflow(remain_df)
    
    global output_dataframe
    output_dataframe = from_dataframe(cash_df)
    
    clear_output()
    update()

def cashflow(df):
    df['cash_out'] = 0
    df['cash_in'] = 0
#     buy_sell_df['remain'] = 0
    buy_num = df['buy_num'].tolist()
    sell_num = df['sell_num'].tolist()
    if_buy = df['b_s'].tolist()
    if_sell = df['s_s'].tolist()
    buy_at = df['buy_at'].tolist()
    sell_at = df['sell_at'].tolist()
    
    cash_out = [b_n*i_b*b_a*-1 for b_n,i_b,b_a in zip(buy_num,if_buy,buy_at)]
    cash_in = [s_n*i_s*s_a for s_n,i_s,s_a in zip(sell_num,if_sell,sell_at)]
    
    df['cash_out'] = cash_out
    df['cash_in'] = cash_in
#     sell_result = [s_n*i_s*-1 for s_n,i_s in zip(sell_num,if_sell)]
#     result = [a+b for a,b in zip(buy_result, sell_result)]  
    
    return df
    

def check_buy_sell(df):
    df['b_s'] = 0
    df['s_s'] = 0
    for index, row in df.iterrows():
        if row['buy_at'] > selected.loc[selected['Date']==index]['Low'].values[0]:
            df.at[index, 'b_s'] = 1
        if row['sell_at'] < selected.loc[selected['Date']==index]['High'].values[0]:
            df.at[index, 's_s'] = 1  
    return df
    
def caculate_remain(buy_sell_df):
    buy_sell_df['remain'] = 0
    buy_num = buy_sell_df['buy_num'].tolist()
    sell_num = buy_sell_df['sell_num'].tolist()
    if_buy = buy_sell_df['b_s'].tolist()
    if_sell = buy_sell_df['s_s'].tolist()
    
    buy_result = [b_n*i_b for b_n,i_b in zip(buy_num,if_buy)]
    sell_result = [s_n*i_s*-1 for s_n,i_s in zip(sell_num,if_sell)]
    result = [a+b for a,b in zip(buy_result, sell_result)]

    remain_list = np.cumsum(result,axis=0)
    buy_sell_df['remain'] = remain_list

#     print(buy_sell_df)
    return buy_sell_df
    
exe_button.on_click(on_button_clicked)

# grid = GridspecLayout(2, 2)
right_box = VBox([buy_num, sell_num, exe_button], layout=Layout(
    display='flex',
    flex_flow='column',
    border='solid .5px',
    align_items='stretch',
    height='200px'
))

global output_dataframe
output_dataframe = sheet()

def update():
    display(AppLayout(header=output_dataframe,
          left_sidebar=date_picker,
          center=right_box,
          right_sidebar=None,
          footer=None))
    
update()
```
